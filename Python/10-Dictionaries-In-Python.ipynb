```python
# Nested Dictionary  --> it means dictionary Inside the Dictionary

students = {

    'student1' :{"name":"aniket" , "age":32},
    'student2' :{"name":"aniket2" , "age":23},
    'student3' :{"name":"aniket3" , "age":22},
    'student4' :{"name":"aniket4" , "age":21},

}

print(students)
```

```python
# Access Nested Dictionary Elements

print(students["student1"]["name"])
print(students["student2"]["age"])

students.items()
```

```python
# Iterating Over nested Dictionary

for student_id , student_info  in students.items():
    print(f"{student_id} : {student_info}")

for key , value in student_info.items():
    print(f"{key} : {value}")
```

```python
#  Dictionary Comprehension

squares = {x : x ** 2 for x in range(5)}
print(squares)
```

```python
# Conditional Dictionary Comprehension

even = {x : x ** 2 for x in range(10) if x % 2 == 0}
print(even)
```

```python
# Practical Example   -> use a dictionary yo count the elements in list

numbers = [1 , 2, 2 , 3 ,3 , 3 , 5 ,6 , 7,8,4 ,4 ]

frequancy = {}
for number in numbers:
    if number in frequancy:
        frequancy[number] += 1
    else:
        frequancy[number] = 1

        print(frequancy)
```

```python
#  Merge Two Dictionaries

dict1 = {"a":1 , "b":2}
dict2 = {"c":3 , "d":4}

merge_dict = {**dict1 , **dict2}   # this will bascially called the keyword argument
print(merge_dict)

#  any numbers of key value pairs are carry it will takes key value pair and it jsut append , inside particular dictionary
```
