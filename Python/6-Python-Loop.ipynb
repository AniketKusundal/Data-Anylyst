```python
#  Loops Control Statements In Python

# 1. Break Statement
# 2. Continue Statement
# 3. Pass Statement
# 4. Nested Loop
```

```python
# 1 . Break Statement
# The break statement in Python is used to terminate the loop prematurely when a certain condition is met


for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# 2. Continue Statement
# The continue statement in Python is used to skip the current iteration of a loop and move to the next iteration when a certain condition is met.



for i in range(10):
    if i % 2 == 0:
        continue
    print(i)
```

```python
# 3. Pass Statement
# The pass statement in Python is a null statement that is used as a placeholder when a statement is syntactically required but no action is needed.

# The pass statment is a null oprator it does nothing

for i in range(5):
    if i == 3:
        print("The Number Is " , i)
        pass
    print(i)
```

```python
# 4 . Nested Loop
# A nested loop is a loop that is contained within another loop. The inner loop is executed completely for each iteration of the outer loop.

for i in range(3):
    for j in range(2):
        print(f"Outer Loop: {i}, Inner Loop: {j}")
```

```python
# Example 1: calculate the sum of the frist n natural number using a while loop 

n = 10
sum = 0
count = 1

while count <= n:
    sum = sum + count
    count = count + 1
print("The Sum Of First N Natural Number is :" , sum)
```
