```python
#  Example 1  Create a To Do list to keep track of task


to_do_list = ["go to gym" , "Eat Helthy Food " , "Clean the house"]

# Adding New Task

to_do_list.append("Go To Office")

# Removing a completed task
to_do_list.remove("Clean the house")

# Checking if the task is list 

if "go to gym" in to_do_list:
    print("Hey don't forgot to go gym")


print(f"Remaning To Do List : - {to_do_list}")


for task in to_do_list:
    print(f"Tasks : {task}")
```

```python
#  Example 2   Organize the Studnets Grade 

grade = [78 , 80 , 99 , 87 , 91 , 92 , 81 , 70] 

# Adding new grade

grade.append(60)


# Calculate The avg grade

avg_grade = sum(grade) / len(grade)
print(f"Avrage Grade -  {avg_grade}")


# finding Height anad lowest grade

highest_grade = max(grade)
print(f"Highest Grade - {highest_grade}")

lowest_grade = min(grade)
print(f"Lowest Grade - {lowest_grade}")
```

```python
# Collecting User Feedback

feedback = ["Very Good" , "Excellent" , "Average" , "Very Satisfied" , "Not To Good"]

# Adding New Feedback

feedback.append("Satisfied")

# count specifivc feedback
positive_feedback = sum(1 for comment in feedback if "Very Goood" in comment.lower() or "Satisfied" in comment.lower())
print(f"Feedback Count : {positive_feedback}" )



# Printing All the feedback 

print("User Feedback")

for comment in feedback:
    print(f" - {comment}")
```
