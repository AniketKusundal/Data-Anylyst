```python
#  Nasted Conditional Statment  --- > we can use the combination one or more if else statment inside  , if ,elif or else statment to create a nested condiitonal statments


# Check Number Is Even Or Odd

num = int(input("Enter The Number : "))

if num > 0:
    print("Number Is Positive")
elif num % 2 == 0:
    print("Number Is Even")
else:      
    print("Number Is Odd")
```

```python
# Conditional Statment Assignment 1 ==> Calculator Program Using the Nested If Else Statment

num1 = float(input("Enter First Number"))
num2 = float(input("Enter Second Number"))


oprator = input("Enter The Operator (+, -, *, /): ")


if oprator == '+':
    result = num1 + num2
elif oprator == '-':
    result = num1 - num2 
elif oprator == '*':
    result = num1 + num2
elif oprator == '/':
    result = num1 / num2
else:
    result = "Invalid Oprator"

print("The Result Is : ", result)

```
