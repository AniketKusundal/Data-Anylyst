```python

number1 = input("Enter The Number 1 : ")
number2 = input("Enter The Number 2 : ")

print(int(number1) + int(number2))
print(int(number1) - int(number2))
print(int(number1) * int(number2))
print(int(number1) / int(number2))
print(int(number1) % int(number2))
print(int(number1) // int(number2))
```
