```python
#  Declaring And Assign Variable 

age = 61
height = 5.8
name = "Aniket"
is_student = True

print("Age Is" , age)
print("Name Is" , height)
print("name Is" , name)
print("Is Student" , is_student)


#  Types oF data in Paython
print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
```

```python
#  Type Conversion In Python   ---> It means changing the data type of the variable

age_str = str(age)
print(type(age_str))    # o/p is string
print(age_str)
```
